```python
y1 = data_1_A1.loc[0]
y2 = data_1_A1.loc[1]
y3 = data_1_A1.loc[2]
y4 = data_1_A1.loc[3]
y5 = data_1_A1.loc[4]
```

```python
from scipy import stats
#当不确定两总体方差是否相等时，应先利用levene检验，检验两总体是否具有方差齐性。
#如果返回结果的p值远大于0.05，那么我们认为两总体具有方差齐性。
p = stats.levene(y1,y2)
p
```

# 两独立样本T检验

```python
#如果返回结果的p值远大于0.05，那么我们认为两总体没有显著性差异。
#ttest_ind(a, b[, axis, equal_var, nan_policy]) #Calculate the T-test for the means of two independent samples of scores
p1 = stats.ttest_ind(y1,y2)
p1
```

结果可以看出来，pvalue=0.014<0.05,说明样本之间的显著性差异较大

```python
#ttest_rel(a, b[, axis, nan_policy]) #Calculate the T-test on TWO RELATED samples of scores, a and b.
p1 = stats.ttest_rel(y1,y2)
p1
```
